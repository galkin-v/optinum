```python
import numpy as np
from optinum.eigen import (power_method, qr_algorithm, compute_spectrum,
                          compute_pseudospectrum, plot_pseudospectrum)
```

```python
print("\n=== Степенной метод ===")
# Тестовая матрица с известными собственными значениями
A = np.array([
    [4, -1],
    [2, 1]
])

lambda_max, eigenvector = power_method(A)
print(f"Наибольшее собственное значение: {lambda_max:.6f}")
print(f"Соответствующий собственный вектор: {eigenvector}")
```

```python
print("\n=== QR-алгоритм ===")
# Симметричная матрица для тестирования
A = np.array([
    [4, -1, 1],
    [-1, 3, -2],
    [1, -2, 3]
])

eigenvalues, eigenvectors = qr_algorithm(A)
print("Собственные значения:")
print(eigenvalues)
print("\nСобственные векторы (по столбцам):")
print(np.array(eigenvectors))
```

```python
print("\n=== Вычисление спектра ===")
# Матрица с комплексными собственными значениями
A = np.array([
    [0, -1],
    [1, 0]
])

spectrum = compute_spectrum(A)
print("Спектр матрицы:")
print(spectrum)
```

```python
print("\n=== Псевдоспектр ===")
# Матрица с интересным псевдоспектром
A = np.array([
    [1, 1, 0],
    [0, 1, 1],
    [0, 0, 1]
])

# Вычисление и отображение псевдоспектра
plot_pseudospectrum(A)
```

```python
print("\n=== Сравнение различных матриц ===")

# Симметричная матрица
A1 = np.array([[2, 1], [1, 2]])

# Верхнетреугольная матрица
A2 = np.array([[2, 1], [0, 3]])

# Матрица с комплексными собственными значениями
A3 = np.array([[0, -1], [1, 0]])

matrices = {"Симметричная": A1, 
            "Верхнетреугольная": A2, 
            "С комплексными с.з.": A3}

for name, matrix in matrices.items():
    print(f"\n{name} матрица:")
    print(matrix)
    spectrum = compute_spectrum(matrix)
    print("Спектр:", spectrum)
```
